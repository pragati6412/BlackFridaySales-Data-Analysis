```python
import pandas as pd
df=pd.read_csv('BlackFriday.csv')
del df['Product_Category_2']
del df['Product_Category_3']
```

```python
df.head()
```

```python
df['Age'].unique()
```

```python
df.groupby('Age').size().plot(kind='bar',figsize=(6,6),title='Product Distribution by Age')
```

```python
df.groupby('Age').sum()['Purchase'].plot(kind='bar',x='Age',figsize=(10,5))
```

```python
lis=[]
for i in df['Age'].unique():
    lis.append([i,(df[df['Age']==i]['Product_ID'].nunique())])
data=pd.DataFrame(lis,columns=['Age','No of Products'])
```

```python
data
```

```python
data.plot(kind='bar',figsize=(6,5),title='Unique Product bought a/c to age')
```

```python
data['No of Products'].sort_values(ascending=False)
```

```python
df.groupby('Age').mean()['Purchase'].plot(kind='bar',figsize=(8,5))
```

```python
df.groupby('Age').mean()['Purchase'].plot(kind='pie',figsize=(6,6),autopct='%.2f')
```

```python
df.groupby('Age').sum()['Purchase'].plot(kind='pie',figsize=(5,5),autopct='%.2f')
```
