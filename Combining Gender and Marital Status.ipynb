```python
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv('BlackFriday.csv')
```

```python
df.head()
```

```python
df1=df[df['Gender']=='M']
sns.countplot(x='Marital_Status',data=df1)
```

```python
sns.countplot(x='Gender',hue='Marital_Status',data=df)
```

```python
l=[]
for i in range(len(df)):
    l.append(df['Gender'][i]+'_'+str(df['Marital_Status'][i]))
```

```python
del df['Product_Category_2']
del df['Product_Category_3']
df['MaritalGender']=l
df.head()
```

```python
df.groupby('MaritalGender').sum()['Purchase'].plot(kind='pie',figsize=(5,5),autopct='%.2f')
```

```python
df.groupby('MaritalGender').mean()['Purchase'].plot(kind='pie',figsize=(5,5),autopct='%.2f')
```

```python
sns.countplot(x='Age',hue='MaritalGender',data=df)
```

```python
sns.countplot(x='City_Category',hue='MaritalGender',data=df)
```
