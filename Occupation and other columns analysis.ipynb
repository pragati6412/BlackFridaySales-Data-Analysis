```python
import pandas as pd
```

```python
df=pd.read_csv('BlackFriday.csv')
del df['Product_Category_2']
del df['Product_Category_3']
```

```python
df.head()
```

```python
import seaborn as sns
```

```python
sns.set(rc={'figure.figsize':(10,5)})
sns.countplot(x='Stay_In_Current_City_Years',data=df)
```

```python
df.groupby('Stay_In_Current_City_Years').size().plot(kind='pie',figsize=(5,5),autopct='%.2f')
```

```python
sns.countplot(x='Stay_In_Current_City_Years',hue='Marital_Status',data=df)
```

```python
sns.countplot(x='Stay_In_Current_City_Years',hue='City_Category',data=df)
```

```python
sns.countplot(x='City_Category',hue='Age',data=df)
```

```python
df.groupby('City_Category').sum()['Purchase'].plot(kind='pie',figsize=(5,5),autopct='%.2f')
```

```python
df.groupby('City_Category').mean()['Purchase'].plot(kind='pie',figsize=(5,5),autopct='%.2f')
```

```python
df.groupby('Occupation').size().plot(kind='pie',figsize=(5,5),autopct='%.2f')
```

```python
df.groupby('Occupation').size().plot(kind='bar',figsize=(10,5))
```

```python
df.groupby('Occupation').sum()['Purchase'].plot(kind='bar',figsize=(10,5))
```

```python
df.groupby('Occupation').mean()['Purchase'].plot(kind='bar',figsize=(10,5))
```

```python
sns.set(rc={'figure.figsize':(10,6)})
sns.countplot(x='Occupation',hue='Marital_Status',data=df)
```

```python
sns.set(rc={'figure.figsize':(10,6)})
sns.countplot(x='Occupation',hue='Gender',data=df)
```

```python
df.groupby('Occupation').nunique()['Product_ID'].plot(kind='bar',figsize=(10,5))
```

```python
df.groupby('Occupation').nunique()['Product_ID'].sort_values().plot(kind='bar',figsize=(10,5))
```
